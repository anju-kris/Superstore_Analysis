```python
import numpy as np
import pandas as pd

# 1. Load the dataset
df = pd.read_csv("superstore_raw.csv")
print("--- Initial Dataset Info ---")
print(df.info())
```

```python

print(df.shape)
```

```python
print(df.head())
```

```python
# 2. Standardize Column Names for PostgreSQL compatibility
# Strips whitespace, lowers text, and replaces spaces/hyphens with underscores
df.columns = (
    df.columns.str.strip().str.lower().str.replace(" ", "_").str.replace("-", "_")
)
```

```python
# 3. Handle Missing & Text Data Anomalies
# Clean text encoding issues in 'product_name' (e.g., fixing the "â€" characters)
if "product_name" in df.columns:
    df["product_name"] = (
        df["product_name"]
        .str.encode("ascii", "ignore")
        .str.decode("utf-8")
        .str.strip()
    )
```

```python
# 4 Standardize postal code (fill missing with '0', force to int then string)
if "postal_code" in df.columns:
    df["postal_code"] = df["postal_code"].fillna(0).astype(int).astype(str)
```

```python
# 5. Remove Duplicate Rows
duplicate_count = df.duplicated().sum()
print(f"\nFound {duplicate_count} duplicate rows.")
if duplicate_count > 0:
    df = df.drop_duplicates().reset_index(drop=True)
```

```python
# 6. Fix Character Encoding Glitches in Product Name
# Replacing the garbled text with standard inch marks (") and fractions
if 'product_name' in df.columns:
    df['product_name'] = df['product_name'].str.replace('â€', '"', regex=False)
    df['product_name'] = df['product_name'].str.replace('â€œ', '"', regex=False)
    df['product_name'] = df['product_name'].str.replace('â€', '"', regex=False)  # Catch-all
    df['product_name'] = df['product_name'].str.replace('Â¾', '3/4', regex=False)
```

```python
# 7. Save the PostgreSQL-ready file
df.to_csv("superstore_final.csv", index=False)

print("\n--- Cleaned Dataset Info ---")
print(df.info())
print(f"\nSuccess! Cleaned file saved as: '{"superstore_final.csv"}'")
```
